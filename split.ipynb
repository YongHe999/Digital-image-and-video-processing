```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
imgA1_1 = cv2.imread('ocr.jpg')
imgA1_2 = cv2.cvtColor(imgA1_1,cv2.COLOR_BGR2GRAY)
fig = plt.gcf()
fig.set_size_inches(20, 20)
plt.imshow(imgA1_2,'gray')
plt.axis("off")
plt.show()
```

```python
def verti(img):
    x,y=img.shape
    flag1=0
    up=[]
    down=[]
    for i in range(x):
        a=img[i,:][img[i,:]<150]
        if len(a)>0 and flag1==0:
            flag1=1
            up.append(i)
        if len(a)==0 and flag1==1:
            flag1=0;
            down.append(i)
    return up,down
```

```python
def hori(img):
    x,y=img.shape
    flag1=0
    left=[]
    right=[]
    for i in range(y):
        a=img[:,i][img[:,i]<150]
        if len(a)>0 and flag1==0:
            flag1=1
            left.append(i)
        if len(a)==0 and flag1==1:
            flag1=0;
            right.append(i)
    return left,right
```

```python
up,down=verti(imgA1_2)
```

```python
x=len(up)
c=0
for i in range(x):
    imgA1_3=imgA1_2[up[i]:down[i],:]
    left,right=hori(imgA1_3)
    y=len(left)
    for j in range(y):
        imgA1_4=imgA1_3[:,left[j]:right[j]]
        if c<100:
            c=c+1
            fig = plt.gcf()
            fig.set_size_inches(20, 10)
            plt.subplot(10,10,c)
            plt.imshow(imgA1_4,'gray')
            plt.axis("off")
plt.show()
```
